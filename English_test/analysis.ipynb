```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.lines import Line2D

import data_loader
```

```python
data = data_loader.data('data')
```

```python
human_verb_type, human_verb_resp, model_line = data.get_verb()
sns.violinplot(x=human_verb_type,y=human_verb_resp,color='white',alpha=0.3,ci=95,capsize=0.05, inner="quartile",palette=['lightblue','orange',])
sns.pointplot(x=human_verb_type,y=human_verb_resp,ci=95,dodge=True,linestyles=[''],capsize=0.05,color='black',scale=0)

i=-0.1
mi = 0
markers = ['b','B','r','R']
mcolors = ['blue','darkred']
for keys, values in model_line.items():
    for j, v in enumerate(values):
        plt.text(i+j*1,v,markers[mi], horizontalalignment='left', size='large', color=mcolors[j], weight='semibold')
    plt.ylim(-0.2,1.3)
    plt.xlim(-0.5,1.5)
    i+=0.05
    mi+=1 
plt.axhline(y=0.5,color='r',linestyle='--')    
legend_elements = [Line2D([0], [0], color='w', lw=0, label='b: BERT-base'),
                   Line2D([0], [0], color='w', lw=0, label='B: BERT-large'),
                   Line2D([0], [0], color='w', lw=0, label='r: RoBERTa-base'),
                   Line2D([0], [0], color='w', lw=0, label='R: RoBERTa-large'),]
plt.legend(handles=legend_elements, bbox_to_anchor=(0.39, 0.32))
plt.ylabel("'for' proportion")
```

```python
article_type, human_article_resp, model_line = data.get_NP()
sns.violinplot(x=article_type,y=human_article_resp,color='white',alpha=0.3,ci=95,capsize=0.05,inner="quartile")
sns.violinplot(x=article_type,y=human_article_resp,alpha=0.3,ci=95,capsize=0.05,inner="quartile",palette=['pink','thistle','lightblue','orange','wheat',])
sns.pointplot(x=article_type,y=human_article_resp,ci=95,dodge=True,linestyles=[''],capsize=0.05,s=0,color='black',scale=0)

markers = ['b','B','r','R']
mcolors = ['crimson','purple','blue','darkred','goldenrod']
i=-0.4
mi = 0
for keys, values in model_line.items():
    for j, v in enumerate(values):
        plt.text(i+j*1,v,markers[mi], horizontalalignment='left', size='large', color=mcolors[j], weight='semibold')
    plt.ylim(-0.2,1.3)
    i+=0.2
    mi+=1
    
legend_elements = [Line2D([0], [0], color='w', lw=0, label='b: BERT-base'),
                   Line2D([0], [0], color='w', lw=0, label='B: BERT-large'),
                   Line2D([0], [0], color='w', lw=0, label='r: RoBERTa-base'),
                   Line2D([0], [0], color='w', lw=0, label='R: RoBERTa-large'),]
plt.axhline(y=0.5,color='r',linestyle='--')  
plt.legend(handles=legend_elements, bbox_to_anchor=(1.01, 1.01))
plt.ylabel("'for' proportion")
```
